# Code

```python
# Text Cleaning

def Preprocessing(task1):
  
  sentences = sent_tokenize(task1)
  sentence_to_words=[]
  for i in sentences[0:]:
   sentence_to_words.extend(i.split())  
  tokens = [ w for w in sentence_to_words if w[0]!='[' and w[-1]!= ']' ]
  import string
  remove = string.punctuation
  import re 
  remove = remove.replace(".", "").replace("-", "").replace(",", "").replace("%", "")
  pattern = r"[{}]".format(re.escape(remove))
  table = str.maketrans('', '', pattern)
  stripped = [w.translate(table) for w in tokens]
  words_to_sentense=' '.join(stripped)
  words_to_sentense
  sentences_new = sent_tokenize(words_to_sentense)
  return sentences_new

# In-built functions

def greedy_decoding (inp_ids,attn_mask):
  greedy_output = model.generate(input_ids=inp_ids, attention_mask=attn_mask, max_length=256)
  Question =  tokenizer.decode(greedy_output[0], skip_special_tokens=True,clean_up_tokenization_spaces=True)
  return Question.strip().capitalize()


def beam_search_decoding (inp_ids,attn_mask):
  beam_output = model.generate(input_ids=inp_ids,
                                attention_mask=attn_mask,
                                max_length=256,
                              num_beams=1,
                              num_return_sequences=1,
                              no_repeat_ngram_size=3,
                              early_stopping=True
                              )
  Questions = [tokenizer.decode(out, skip_special_tokens=True, clean_up_tokenization_spaces=True) for out in
              beam_output]
  return [Question.strip().capitalize() for Question in Questions]


# Generate Question & Context

def Questions_Context(task2):
    max_len = 512
    context_list=[]
    question_list=[]

    for i in task2: 
      encoding = tokenizer.encode_plus(i, return_tensors="pt", add_special_tokens=True,max_length=max_len, truncation=True,padding=True)
      input_ids, attention_masks = encoding["input_ids"], encoding["attention_mask"]
      context_list.append(i)
      question_append = beam_search_decoding(input_ids,attention_masks) 
      for out in question_append:
        question_list.append(out)
    return question_list,context_list

#Remove Duplicates
def Duplicate_removing(task3):
  new_questions = task3
  new_questions4 = []

  [new_questions4.append(x) for x in new_questions if x not in new_questions4]
  return new_questions4

# Prediction of Answers

def predict(question, passage):
  sequence = tokenizer_2.encode_plus(question, passage, return_tensors="pt")['input_ids']

  logits = model_2(sequence)[0]
  probabilities = torch.softmax(logits, dim=1).detach().cpu().tolist()[0]
  proba_yes = round(probabilities[1], 2)
  proba_no = round(probabilities[0], 2)
  if proba_yes>0.50:
    return True
  else:
    return False 


def Answer_prediction(question_list, context_list):
  answers=[]
  for i, j in zip(question_list, context_list ):
      answers.append(predict(i, j))
  return answers

# Define Questions

def Generate_questions(new_questions4):
  Ques_Ans = []
  for i in new_questions4:
    Ques_Ans.append({"question":i})
  return Ques_Ans

# Ranking of Questions

def Ranking(question_list, context_list):
  de=[]
  for i, j in zip(question_list, context_list ):
    d = question_answerer(question=i, context=j)
    e=d.get("score")
    e_updated=1-float(e)
    accuracy_score="%.2f" % round(e_updated*100, 2)
    de.append(accuracy_score)
  return de

# Get Context, Rank,Answer & Paragraph no.

def Final_dictionary(Ques_Ans,context_list,de,answers):
  for option_index, option in enumerate(Ques_Ans):
      option["context"] = context_list[option_index] # context_list
      option["rank"] = de[option_index]
      option["answer"] = answers[option_index]  
      # option['Paragraph_no']= int(j+1)
  return Ques_Ans

# Sorting as per rank
def Sorting(Ques_Ans):
  final_generated_questions = sorted(Ques_Ans, key=lambda i:float(i['rank']),reverse=True)
  return final_generated_questions

# Extract single list from nested list
def flatten(list_of_lists):
    return list(itertools.chain.from_iterable(list_of_lists))

# define user required questions
generate_ques=int(number_of_question)

# Pipeline

def Pipeline(text):
  a=Preprocessing(text)                    # Text cleaning & pre-processing
  b=Questions_Context(a)                   # generate questions from context
  c=Duplicate_removing(b)                  # remove duplicate questions & context
  d= Answer_prediction(c[0],c[1])          # Answer prediction fron questions & context
  e=Generate_questions(c[0])               # Getting questions in dictionary
  f=Ranking(c[0],c[1])                     # find rank from question list & context list
  g=Final_dictionary(e,c[1],f,d)           # Final dictionary consisting question, context, rank & answer
  return g
```

```python
import time
import torch
from IPython.utils.path import ENOLINK
from transformers import T5ForConditionalGeneration,T5Tokenizer
from transformers import pipeline
from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
)
import json
```

```python
def regenerate_answer(question,data_text):

    regenerate_answer_model = AutoModelForSequenceClassification.from_pretrained("shahrukhx01/roberta-base-boolq")
    tokenizer_regenerate = AutoTokenizer.from_pretrained("shahrukhx01/roberta-base-boolq")

    sequence = tokenizer_regenerate.encode_plus(question,data_text, return_tensors="pt")['input_ids']
    logits = regenerate_answer_model(sequence)[0]
    probabilities = torch.softmax(logits, dim=1).detach().cpu().tolist()[0]
    proba_yes = round(probabilities[1], 2)
    proba_no = round(probabilities[0], 2)
    if proba_yes>0.5:
      return True
    else:
      return False

def regenerte_quest(original_id,original_context,data_text, data_answer,data_old_quest,data_question_no,data_question_from,data_question_type):
  
  model = T5ForConditionalGeneration.from_pretrained('ramsrigouthamg/t5_boolean_questions')
  tokenizer = T5Tokenizer.from_pretrained('t5-base')
  

  def topkp_decoding (inp_ids,attn_mask):
    topkp_output = model.generate(input_ids=inp_ids,
                                  attention_mask=attn_mask,
                                  max_length=256,
                                do_sample=True,
                                top_k=40,
                                top_p=0.80,
                                num_return_sequences=1,
                                  no_repeat_ngram_size=2,
                                  early_stopping=True
                                )
    Questions = [tokenizer.decode(out, skip_special_tokens=True,clean_up_tokenization_spaces=True) for out in topkp_output]
    return [Question.strip().capitalize() for Question in Questions]

  max_len = 512
  encoding = tokenizer.encode_plus(data_text, return_tensors="pt", add_special_tokens=True,max_length=max_len, truncation=True,padding=True)
  input_ids, attention_masks = encoding["input_ids"], encoding["attention_mask"]
  question_append_topk = topkp_decoding(input_ids,attention_masks)
  for i in question_append_topk:
    regenerate_question = i

  question_answerer = pipeline("question-answering")
  d = question_answerer(question=question_append_topk, context=data_text)
  e=d.get("score")
  e_updated=1-float(e)
  rank_regenerate="%.2f" % round(e_updated*100, 2)
    
  generted_quest_is=[]
  generted_quest_is.append({"file_id":original_id,"rege_question":regenerate_question, "answer":None,"rank" :rank_regenerate , "context":original_context,"old_question":data_old_quest, "question_id" : data_question_no, "question_from":data_question_from,"question_type":data_question_type, "status":"Done"})
 
  for i in generted_quest_is:
    question = i["rege_question"]
    final_question_is = regenerate_answer(question,data_text)
    i["answer"] = final_question_is

  url="https://generate-questions.devbyopeneyes.com/api/reGenerateQuestions"
  headers = {'Content-Type':'application/json','Accept':'application/json'}
  status = requests.post(url,headers=headers,data=json.dumps(i))

  return status, i
```
