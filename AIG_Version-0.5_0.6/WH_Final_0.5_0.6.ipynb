# Code

```python
# Text Cleaning

def Preprocessing(task1):
  
  sentences = sent_tokenize(task1)
  sentence_to_words=[]
  for i in sentences[0:]:
   sentence_to_words.extend(i.split())  
  tokens = [ w for w in sentence_to_words if w[0]!='[' and w[-1]!= ']' ]
  import string
  remove = string.punctuation
  import re 
  remove = remove.replace(".", "").replace("-", "").replace(",", "")
  pattern = r"[{}]".format(re.escape(remove))
  table = str.maketrans('', '', pattern)
  stripped = [w.translate(table) for w in tokens]
  words_to_sentense=' '.join(stripped)
  return words_to_sentense

# get_question

def get_question(sentence, answer): 

  text = "context: {} answer: {}".format(sentence,answer)
  max_len = 256
  encoding = tknizer.encode_plus(text,max_length=max_len, pad_to_max_length=False,truncation=True, return_tensors="pt")
  input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]
  outs = mdl.generate(input_ids=input_ids,
                                  attention_mask=attention_mask,
                                  early_stopping=True,
                                  num_beams=5,
                                  num_return_sequences=1,
                                  no_repeat_ngram_size=2,
                                  max_length=300)
  dec = [tknizer.decode(ids,skip_special_tokens=True) for ids in outs]
  Question = dec[0].replace("question:","")
  Question= Question.strip()
  return Question

# In_built functions

def get_embedding(doc):
  # txt = '[CLS] ' + doc + ' [SEP]'
  tokens = bert_tokenizer.tokenize(txt)
  token_idx = bert_tokenizer.convert_tokens_to_ids(tokens)
  segment_ids = [1] * len(tokens)

  torch_token = torch.tensor([token_idx])
  torch_segment = torch.tensor([segment_ids])

  return bert_model(torch_token, torch_segment)[-1].detach().numpy()

def get_pos(context):
  doc = nlp(context)
  docs = [d.pos_ for d in doc]
  return docs, context.split()

def get_sent(context):
  doc = nlp(context)
  return list(doc.sents)

def get_vector(doc):
  stop_words = "english"
  n_gram_range = (1,1)
  df = CountVectorizer(ngram_range = n_gram_range, stop_words = stop_words).fit([doc])
  return df.get_feature_names()


def get_key_words(context, module_type = 't'):
  keywords = []
  top_n = 5
  for txt in get_sent(context):
    keywd = get_vector(str(txt))
    if module_type == 't':
      doc_embedding = get_embedding(str(txt))
      print("--doc_embedding--",doc_embedding)
      keywd_embedding = get_embedding(' '.join(keywd))
    else:
      doc_embedding = model.encode([str(txt)])
      keywd_embedding = model.encode(keywd)
    
    distances = cosine_similarity(doc_embedding, keywd_embedding)
    keywords += [(keywd[index], str(txt)) for index in distances.argsort()[0][-top_n:]]

  return keywords

# Duplicate removing

def remove_duplication(data_questions_arr):
  newQuestions = []
  for i in data_questions_arr:
    if i not in newQuestions:
      newQuestions.append(i)
  return newQuestions


# Ranking of questions

def get_score(data_new_questions,data_context):
    question_answerer = pipeline("question-answering")
    context = data_context
    question_score = []
    for i in data_new_questions:
        final_score_list = question_answerer(question=i['question'], context=context)
        final_score = final_score_list.get("score")
        final_score ="%.2f" % round(final_score*100, 2)
        question_score.append(final_score)
    for option_index, option in enumerate(data_new_questions):
        option["rank"] = question_score[option_index]
    return data_new_questions
    
def correct_latter_question_context(question_context_data):
    import re
    remove_quots =re.sub("[”]","",question_context_data)
    remove_space = remove_quots.strip()
    new_sentence = remove_space.capitalize()
    return new_sentence

def correct_list(data):
    temp_list = []
    for i in data:
        i["question"] = correct_latter_question_context(i["question"])
        i["context"] = correct_latter_question_context(i["context"])
        i["answer"] = correct_latter_question_context(i["answer"])
        temp_list.append(i)
    return temp_list

#  number of WH questions made with context & Answer

def WH_question(task2):
  sentence_count = len(task2.split("."))    
  word_count = sum([i.strip(string.punctuation).isalpha() for i in str(task2).split()])
  if sentence_count > 2 and word_count > 10:
    question_number_is = sentence_count + 2
  elif sentence_count == 1 and word_count > 7:
    question_number_is = sentence_count + 1
  else:
    question_number_is = 1
  questions_arr = []
  for answer, context in get_key_words(task2, 'st'):
    questions_arr.append({"question":get_question(context, answer),"context":context,"answer":answer})
    if len(questions_arr) == question_number_is:
      break
  return questions_arr

def Sorting(final_required_data):
  final_data = sorted(final_required_data, key=lambda i: float(i['rank']),reverse=True)
  return final_data


  
# Extract single list from nested list
def flatten(list_of_lists):
    return list(itertools.chain.from_iterable(list_of_lists))

    
# Pipeline

def Pipeline(text):
  a=Preprocessing(text)      # text cleaning & preprocessing by paragraph
  b=WH_question(a)           # WH-questions generated with answer & context
  c=remove_duplication(b)    # Remove duplicates
  d=get_score(c,a)           # Ranking
  e=correct_list(d)          # Final questions with capitalization 
  return e


```
